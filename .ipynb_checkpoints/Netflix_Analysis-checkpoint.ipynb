# üé¨ Netflix Viewing Habits Analysis

## üìç Objective

This project analyzes Netflix content data to uncover patterns in content type, genre distribution, release trends, and country-wise offerings. By performing Exploratory Data Analysis (EDA), we aim to answer:

- What types of content are most common on Netflix?
- How has the content volume changed over time?
- What are the most popular genres on the platform?
- Which countries contribute the most content?

## üì¶ Dataset

The dataset is sourced from Kaggle and includes metadata about Netflix titles available as of 2021, including information like title, type, cast, country, release year, genre, and more.

```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('netflix_titles.csv')

# Display the first few rows
df.head()
```

```python
# Basic info about the dataset
df.info()
```

```python
# Checking for missing values
df.isnull().sum()
```

```python
# Unique values in each column
df.nunique()
```

## üîç Data Cleaning

```python
# Fill 'country' column safely without chained assignment
df['country'] = df['country'].fillna('Unknown')

# Strip leading/trailing whitespaces from all object columns
df = df.apply(lambda x: x.str.strip() if x.dtype == 'object' else x)
```

```python
# Drop rows with missing 'date_added'
df = df.dropna(subset=['date_added'])

# Now convert to datetime (no need to manually specify format)
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
```

## üìä Content Type Distribution

Let‚Äôs see how many Movies vs TV Shows are on Netflix.


```python
sns.countplot(data=df, x='type', palette='Set2')
plt.title("Distribution of Content Types on Netflix")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

üìå Insight: The dataset has more Movies than TV Shows. This indicates that Netflix primarily focuses on movies in terms of quantity.

## üåç Top 10 Content Producing Countries

This chart shows which countries have produced the most content on Netflix.


```python
top_countries = df['country'].value_counts().head(10)


top_countries.plot(kind='barh', color='skyblue')
plt.title("Top 10 Content Producing Countries")
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.gca().invert_yaxis()
plt.show()
```

## üìÖ Content Added Over the Years

We analyze how Netflix's content library has grown over time.

```python
df_year = df['year_added'].value_counts().sort_index()

plt.figure(figsize=(12,6))
sns.lineplot(x=df_year.index, y=df_year.values, marker='o', linewidth=2)
plt.title("Content Added on Netflix Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.grid(True)
plt.show()
```

üìå Insight: There was a significant growth in content additions from 2015 to 2019. The peak in 2019 may reflect Netflix's expansion strategy before the pandemic.


## üéØ Content Ratings Distribution

What are the most common content ratings like TV-MA, PG, R, etc.?


```python
plt.figure(figsize=(10,5))
sns.countplot(y='rating', data=df, order=df['rating'].value_counts().index, palette='mako')
plt.title("Content Ratings Distribution")
plt.xlabel("Count")
plt.ylabel("Rating")
plt.show()
```

## üé≠ Most Common Genres

We split and explode the `listed_in` column to identify the most frequent genres across the platform.


```python
# Split and explode genres
df_genres = df.assign(genres=df['listed_in'].str.split(',')).explode('genres')
df_genres['genres'] = df_genres['genres'].str.strip()

top_genres = df_genres['genres'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_genres.values, y=top_genres.index, palette='viridis')
plt.title("Top 10 Most Common Genres on Netflix")
plt.xlabel("Count")
plt.ylabel("Genre")
plt.show()
```

üìå Insight: The most common genres are Drama, International Movies, and Comedy, indicating a global and versatile content strategy targeting broad viewer interests.


```python
# Extract release year from 'date_added'
df['year_added'] = pd.DatetimeIndex(df['date_added']).year

# Drop missing years
df_years = df.dropna(subset=['year_added'])

# Plot number of titles added per year
plt.figure(figsize=(10,6))
sns.countplot(data=df_years, x='year_added', order=sorted(df_years['year_added'].unique()), palette='coolwarm')
plt.title("Number of Titles Added to Netflix Per Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

## üìå Conclusion

- The majority of Netflix content is Movies.
- Most content is from the United States, followed by India and the UK.
- The most frequent genres are Dramas, Comedies, and Documentaries.
- Content addition peaked in 2019-2020, possibly due to the pandemic.
- A significant portion of content is intended for a mature audience (TV-MA, R).

This analysis provides insights into how Netflix is expanding its content library and audience reach.
